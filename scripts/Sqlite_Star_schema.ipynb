```python
from sqlalchemy import create_engine
import pandas as pd

import warnings 
warnings.filterwarnings("ignore")
```

```python
engine= create_engine("sqlite:////Users/himanshunautiyal/Desktop/untitled folder/Project/sql/mutual_fund.db")
```

```python
files = {
    "cleaned_fund_master": "/Users/himanshunautiyal/Desktop/untitled folder/Project/data/processed/01_fund_master.csv",
    "cleaned_nav_history" : "/Users/himanshunautiyal/Desktop/untitled folder/Project/data/processed/02_cleaned_nav_history.csv",
    "cleaned_sheme_performance" : "/Users/himanshunautiyal/Desktop/untitled folder/Project/data/processed/07_clean_scheme_performance.csv",
    "cleaned_investor_transaction" : "/Users/himanshunautiyal/Desktop/untitled folder/Project/data/processed/08_cleaned_investor_transaction.csv"
}
```

```python
for table_name, path in files.items():
    df = pd.read_csv(path)

    df.to_sql(table_name,engine,if_exists="replace",index=False)

    sql_table_rcount=pd.read_sql(f"select count(*) as count from {table_name}",engine)["count"]

    print(table_name,
          "\nSource data rows count :", len(df),
          "\nNo. of rows in database table :", sql_table_rcount,
          )
    print("-"*50)
```
