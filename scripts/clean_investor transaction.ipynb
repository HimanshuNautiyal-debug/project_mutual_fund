```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
investor_transactions = pd.read_csv("/Users/himanshunautiyal/Desktop/untitled folder/Project/data/raw/08_investor_transactions.csv")
investor_transactions.head(5)
```

```python
print("Shape:", investor_transactions.shape)
print("Data Types:\n",investor_transactions.dtypes)
```

```python
investor_transactions["transaction_date"] = pd.to_datetime(investor_transactions["transaction_date"],errors="coerce")
investor_transactions.dtypes
```

```python
# checking for unique values in "transaction type"
investor_transactions["transaction_type"].unique()
```

```python
# Standardise transaction_type 
standard_val={
    "sip" : "SIP",
    "Sip" : "SIP",
    "SIP" : "SIP",
    'Standard Investment Plan' : "SIP",
    "lumpsum" : "Lumpsum",
    "Lumpsum" : "Lumpsum",
    "Lump sum" : "Lumpsum",
    "lump sum" : "Lumpsum",
    "redemption" : "Redemption",
    "Redemption" : "Redemption"
    }

investor_transactions["transaction_type"]=investor_transactions["transaction_type"].astype(str).str.strip().replace(standard_val)
```

```python
invalid_amounts=investor_transactions[investor_transactions["amount_inr"]<=0]
print("Total number of invalid amount :",len(invalid_amounts))
```

```python
invalid_kyc=investor_transactions[~investor_transactions["kyc_status"].isin(['Verified', 'Pending'])]
print("Total number of Invalid kyc status: ",len(invalid_kyc))
```
