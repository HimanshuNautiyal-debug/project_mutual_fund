```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
nav_history = pd.read_csv("/Users/himanshunautiyal/Desktop/untitled folder/Project/data/raw/02_nav_history.csv")
nav_history.head(5)
```

```python
#Checking the data types of the column in the DataFrame
print("Shape:", nav_history.shape)
print("Data Types:\n",nav_history.dtypes)
```

```python
#Changing the Data Type of the date column to datetime format
nav_history["date"] = pd.to_datetime(nav_history["date"],errors='coerce')
nav_history.head(25)
```

```python
# Checking Dublicates and droping them if any
nav_history.duplicated().sum() #there are no dublicates in the data
```

```python
#Validate
invalid_nav=nav_history[nav_history["nav"]<0]
print("Total number od invalid nav:",len(invalid_nav))
# There are no invalid nav values in the data.
```

```python
#Forward fill missing nav for weekends and holidays
nav_history["nav"]=nav_history.groupby(["amfi_code"])["nav"].ffill()
```

```python
nav_history.to_csv("/Users/himanshunautiyal/Desktop/untitled folder/Project/data/processed/02_cleaned_nav_history.csv",index=False)
```
