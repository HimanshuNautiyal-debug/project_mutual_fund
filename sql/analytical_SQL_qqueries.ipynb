```python
top_5_funda_by_aum = text("""
             select fund_name, sum(aum) as total_aum from fact_aum fa
             join dim_fund df on fa.find_id = df.fund_id
             group by fund_name 
             order by total_aum desc
             limit 5;
             """)
```

```python
sip_yoy_growth = text(""" 
                      select year, sum(amount) from fact_transaction ft 
                      join dim_date dd on ft.date_id = dd.date_id
                      where transaction_type = "SIP"
                      group by year 
                      """)
```
