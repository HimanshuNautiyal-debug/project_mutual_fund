### Data Directory

## Overview
- This data base folls a Star Schema Design consisting of two Dimensional Table and Four Fact Tables.

### Dimensional Tables:

### dim_fund:
- Source: 01_fund_master.csv

|Column_Name   |   Data_Type  |                     Business_Definition                           |
|--------------|--------------|-------------------------------------------------------------------|
|amfi_code     |     TEXT     |Uniqie AMFI Scheme code identifying the mutual fund and Primary Key|
|fund_name     |     TEXT     |Offecial name of the mutual fund                                   |
|category      |     TEXT     |Mutual Fund Category                                               | 
|fund_house    |     TEXT     |Asset Management Company(AMC) managing the scheme                  |


### dim_date
- Source: Derived from transaction and nav dates

| Column_Name    | Data_Type | Business_Definition               |
|----------------|-----------|-----------------------------------|
|  date_id       |  INTEGER  |  Primary Key for date dimension   |
|  full_date     |  INTEGER  |  Complete calender sate           |
|  year          |  INTEGER  |  Calender Year                    |    
|  quarter       |  INTEGER  |  Calender Quarter (1-4)           |
|  month         |  INTEGER  |  Month Number(1-12)               |
|  day           |  INTEGER  |  Day of month                     | 


## Fact Tables

### fact_nav
- Source=02_nav_history.csv

| Column_Name    | Data_Type | Business_Definition               |
|----------------|-----------|-----------------------------------|
|nav_id|INTEGER|Unique record identifier for NAV entry|
|fund_id|INTEGER|Foreign key referencing dim_fund|
|date_id|INTEGER|Foreign key referencing dim_date|
|nav|INTEGER|Net Asset Value of the scheme on the given date|

### fact_transactions
- Source: 08_investor_transactions.csv

|Column_Name|Data_Type|Business Definition|
|-----------|---------|-------------------|
|transaction_id|INTEGER|	Unique transaction identifier|
|fund_key|INTEGER|Foreign key referencing dim_fund|
|date_key|INTEGER|Foreign key referencing dim_date|
|investor_id|TEXT|Unique investor identifier|
|transaction_type|TEXT|Transaction category (SIP, Lumpsum, Redemption)|
|amount|REAL|Transaction amount in INR|

- Amount must be greater than 0.
- Transaction type must be SIP, Lumpsum, or Redemption.

### fact_performance
- Source: 07_scheme_performance.csv

| Column_Name  | Data_Type  |Business_Definition|
|--------------|----------|------------------|
|performance_id|INTEGER|Unique performance record identifier|
|amfi_code.    |INTEGER|Foreign key referencing dim_fund|
|date_id       | INTEGER|Foreign key referencing dim_date|
|return_1y     |REAL|One-year scheme return percentage|
|return_3y.    | REAL|Three-year scheme return percentage|
|return_5y  |    REAL|Five-year scheme return percentage|
|expense_ratio|REAL|Annual expense ratio charged by the fund|

- Return values must be numeric.
- Expense ratio should typically range between 0.1% and 2.5%.

### fact_aum
- Source: 03_aum_data.csv

| Column_Name  | Data_Type  |Business_Definition|
|--------------|----------|------------------|
|aum_id	|INTEGER	|Unique AUM record identifier|
|fund_key	|INTEGER	|Foreign key referencing dim_fund|
|date_key	|INTEGER	|Foreign key referencing dim_date|
|aum	|REAL	|Assets Under Management (AUM) in INR|

- AUM must be non-negative.
